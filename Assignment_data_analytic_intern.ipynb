First we will import necessary Libraries

### Selecting 10 random companies to determine Intrinsic value assuming 3 cases scenario of growth (g)


```python
df_intrinsic = df[['Name', 'Sub-Sector', 'Market Cap', 'Earnings Per Share']]
df_intrinsic_value = df_intrinsic.sample(10, axis = 0)
df_intrinsic_value
```

### Finding Intrinsic value of 10 randomly selected company assuming 3 Cases for g (Growth) : Good (15% <br> Growth) ; Bad (-5% Growth); Best (25%Growth)

V : Intrinsic Value<br>
EPS : The Company’s last 12 month earnings per share<br>
8.5 : The constant represents the appropriate P-E ratio for a no-growth
      company as proposed by Graham.<br>
g : The company's long-term (five years) earnings growth estimate<br>
6 : The average Return of FDs (6%)<br>
Y : The current yield on AAA corporate bonds (Current yeild on AAA corporate bond as of 2 sept 2022 is **7.253** (Source <a>"http://www.worldgovernmentbonds.com/bond-historical-data/india" </a>)

#### The Data type in "Earnings Per Share" column is in object type, casting it into float for mathematical operation purpose

```python
df_intrinsic_value['Earnings Per Share'] = df_intrinsic_value['Earnings Per Share'].astype(float)
```

```python
df_intrinsic_value.info()
```

#### Calculating the Intrinsic value for 10 selected companies based on 3 cases of growth scenario

```python
# Calculating Intrinsic Value for 10 selected companies with 15% growth
df_intrinsic_value['Intrinsic Value with 15% Growth'] = df_intrinsic_value['Earnings Per Share'] * ((8.5 + 2*15)*6)/7.253
# Calculating Intrinsic Value for 10 selected companies with -5% growth
df_intrinsic_value['Intrinsic Value with -5% Growth'] = df_intrinsic_value['Earnings Per Share'] * ((8.5 + 2*(-5))*6)/7.253
# Calculating Intrinsic Value for 10 selected companies with 25% growth
df_intrinsic_value['Intrinsic Value with 25% Growth'] = df_intrinsic_value['Earnings Per Share'] * ((8.5 + 2* 25)*6)/7.253
```

```python
df_intrinsic_value
```

## Plotting Line chart of the selected 10 companies asuming 3 cases of growth w.r.t. Market capital

```python
from matplotlib import style
plt.figure(figsize =(30, 20))
style.use('ggplot')
plt.plot( df_intrinsic_value['Name'], df_intrinsic_value['Intrinsic Value with 15% Growth'], label = 'Intrinsic Value with 15% Growth', marker ='o', markersize = 10, linewidth = 5)
plt.plot( df_intrinsic_value['Name'], df_intrinsic_value['Intrinsic Value with -5% Growth'], label = 'Intrinsic Value with -5% Growth',marker ='o', markersize = 10, linewidth = 5)
plt.plot( df_intrinsic_value['Name'], df_intrinsic_value['Intrinsic Value with 25% Growth'], label = 'Intrinsic Value with 25% Growth', marker ='o', markersize = 10, linewidth = 5)
plt.xlabel('Name of companies', fontsize = 25)
plt.ylabel('Intrinsic Growth assuming 3 case scenario', fontsize = 25)
plt.title('Intrinsic Growth of 10 selected companies', fontsize = 30)
plt.legend(fontsize = 16)
```

```python
plt.figure(figsize =(30, 20))
style.use('ggplot')
plt.plot( df_intrinsic_value['Name'], df_intrinsic_value['Intrinsic Value with 15% Growth'], label = 'Intrinsic Value with 15% Growth', marker ='o', markersize = 10, linewidth = 5)
plt.plot( df_intrinsic_value['Name'], df_intrinsic_value['Market Cap'], label = 'Market Cap', linewidth = 5, color = 'g')
plt.xlabel('Name of companies', fontsize = 25)
plt.ylabel('Intrinsic Growth against Market Capital', fontsize = 25)
plt.title('Intrinsic Growth of 10 selected companies  with 15% Growth', fontsize = 30)
plt.legend(fontsize = 16)
```
